### Good vs Bad Pandas Practices

**Example 1**

- Convert the column `name` to UPPER case

```python
### Bad approach - using apply 
users['name_upper'] = users.apply(lambda row: row['name'].upper(), axis=1) 
```

```python
users
```

```python
### Ideal approach - using vectorized string operations
users['name_upper'] = users['name'].str.upper()
```

```python
users
```

**Example 2**

- Create a new column that displays the age of the users in months

```python
### Bad approach - using apply
users['age_in_months'] = users.apply(lambda row: row['age']*12, axis=1)
```

```python
users
```

```python
## Ideal approach
users['age_in_months'] = users['age']*12
```

```python
users
```

**Example 3**

- Create a new user_id for each user combining the first two letters of their name and their age. For example, the new user_id of `Alice Kim` with age `58` would be `Al58`.

```python
### Bad approach - using iterrows()

new_userids = [] ##Create an empty list to store the new userids

for index, row in users.iterrows(): 
    new_userids.append(row['name'][0:2] + str(row['age'])) ## Append each user's new userid
    
new_userids
```

```python
### Create a new column 'new_userid' in the DataFrame using the previously created list
users['new_userid'] = new_userids
```

```python
users
```

```python
### Ideal approach -  using vectorized strings
users['new_userid'] = users['name'].str[0:2] + users['age'].astype(str)
```
